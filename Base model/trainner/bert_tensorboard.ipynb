```python
import tensorflow as tf
import matplotlib.pyplot as plt

LOSS = "loss"
EVAL_CCURACY = "eval_accuracy"
LR = 'lr'
EVAL_LOSS = "eval_loss"
tf_event_file = "events.out.tfevents.1650370937.8e3916681d9f"

summary = tf.compat.v1.train.summary_iterator(tf_event_file)
```

```python

loss_list = []
accuracy_list = []
lr_list = []
eval_loss_list = []
for e in tf.compat.v1.train.summary_iterator(tf_event_file):
    for v in e.summary.value:
        if v.tag == LOSS:
            loss_list.append(v.simple_value)
        elif v.tag == EVAL_CCURACY:
            accuracy_list.append(v.simple_value)
        elif v.tag == LR:
            lr_list.append(v.simple_value)
        elif v.tag == EVAL_LOSS:
            eval_loss_list.append(v.simple_value)

plt.plot(loss_list)
plt.grid()
plt.show()
plt.savefig('loss.png')
plt.close()

plt.plot(accuracy_list)
plt.grid()
plt.show()
plt.savefig('eval accuracy.png')
plt.close()
plt.plot(eval_loss_list)
plt.grid()
plt.show()
plt.savefig('eval loss.png')
plt.close()
```
